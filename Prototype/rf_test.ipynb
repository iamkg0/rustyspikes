```python
from rustyspikes import *
```

```python
def run(I=0, time=100):
    model = SNNModel()
    neu_0 = DirectNeuron(I=I)
    neu_1 = Izhikevich()
    syn = Synapse(neu_0, neu_1)
    model.add_neuron(neu_0)
    model.add_neuron(neu_1)
    model.add_synapse(syn)
    model.reload_graph()
    model.define_output(1)
    time = 100
    t = np.arange(int(time / res)) * .1
    for i in t:
        model.tick()
        timing = model.get_response(just_timings=True)
        if timing[0]:
            print(f'{round(i, 5)} ms with I = {round(I, 5)}')
            return round(i, 5), round(I, 5)

Is = []
sps = []
for i in range(5000):
    a = run(I=i*0.01, time=100)
    if a:
        Is.append(a[1])
        sps.append(a[0])
    else:
        pass
```

```python
plt.plot(Is, sps)
```

```python
plt.plot(np.array(sps)/np.array(Is))
for i in np.array(sps)/np.array(Is):
    print(i)
```
